## Task 1

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df_json = pd.read_json("top500.json", orient="index")
df_json.head()
```

```python
df_json["top_1"] = df_json["#1"].apply(lambda x: x[0])
df_json.head()
```

```python
df_json.tail()
```

```python
df_json.reset_index(inplace=True)
df_json.rename(columns={"index": "date"}, inplace=True)
df_json.head()
```

```python
df_top1_json = (
    df_json[
        ["date", 'top_1']
    ].append(
        pd.DataFrame(
            np.nan, 
            index=[i for i in range(6)], 
            columns=["date", 'top_1']
        ), 
        ignore_index=True
    )
)
df_top1_json.head()
```

```python
df_top1_json.shape
```

```python
df_top1_json.tail(10)
```

```python
df_top1_json.loc[64, "date"] = pd.to_datetime('2025-6-1')
df_top1_json.tail(10)
```

```python
df_top1_json["date"] = pd.to_datetime(df_top1_json["date"])
df_top1_json
```

```python
df_top1_json['month'] = df_top1_json["date"].dt.month
df_top1_json.loc[57:64,'month'] = [11, 6] * 4
df_top1_json['month'] = df_top1_json['month'].astype(int)
df_top1_json['year'] = df_top1_json["date"].dt.year.interpolate().astype(int)
df_top1_json['day'] = 1
df_top1_json
```

```python
df_top1_json["date"] = (
    df_top1_json["date"]
    .fillna(
        pd.to_datetime(
            df_top1_json['year']
            .astype(str) 
            + '-' 
            + df_top1_json['month'].astype(str) 
            + '-' 
            + df_top1_json['day'].astype(str)))
)

df_top1_json
```

```python
df_top1_json['days_from_start'] = (df_top1_json["date"] - df_top1_json["date"][0]).dt.days
df_top1_json
```

```python
train = df_top1_json[['days_from_start', 'top_1']].dropna()
train['log_target'] = np.log(train['top_1'])
train.head()
```

```python
lr = LinearRegression()
lr.fit(np.array(train['days_from_start']).reshape(-1, 1), train['log_target'])
```

```python
preds = np.exp(
    lr.predict(
        np.array(
            df_top1_json[
                df_top1_json['top_1'].isna()
            ]['days_from_start']
        ).reshape(-1, 1)
    )
)
preds
```

```python
df_top1_json.tail(10)
```

```python
df_top1_json.loc[59:64, 'top_1'] = preds
df_top1_json
```

```python
df_top1_json.tail(20)
```

```python
df_top1_json.to_csv('answer_top_1.csv',index=False)
```
